```python
import tiktoken
import random
```

```python
text: str

with open("./corpora/sherlock.txt", mode="r") as f:
    text = f.read()
```

```python
tokenizer = tiktoken.get_encoding("gpt2")
```

```python
ids = tokenizer.encode(text)
mi, ma = min(ids), max(ids)
```

```python
rand_ids = [random.randint(mi, ma) for _ in range(100)]
```

```python
tokenizer.decode(rand_ids).split()[:100]
```

```python
vocab = [tokenizer.decode([i]) for i in range(tokenizer.n_vocab)]
vocab[:128]
```

```python
for i, t in enumerate(sorted(vocab, key = lambda t : len(t))):
    if len(t) >1:
        print(i, t.replace(" ", "_"))
    if i > 1000:
        break
```
